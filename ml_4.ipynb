```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
ds = pd.read_csv('TeluguMovies_dataset.csv')
ds
```

```python
ds.head(20)
```

```python
col = ds.columns
col
```

```python
plt.figure(figsize=(6,4))
plt.hist(ds["Rating"], bins=10, color="skyblue", edgecolor="black")

plt.title("Histogram of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(ds["Rating"], bins=10, kde=True)

plt.title("Histogram of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(ds["Runtime"], ds["Rating"], color="red")

plt.title("Runtime vs Rating")
plt.xlabel("Runtime (Minutes)")
plt.ylabel("Rating")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Runtime", y="Rating", data=ds)

plt.title("Runtime vs Rating")
plt.xlabel("Runtime (Minutes)")
plt.ylabel("Rating")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(ds["Runtime"])

plt.title("Box Plot of Runtime")
plt.ylabel("Runtime (Minutes)")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=ds["Runtime"])

plt.title("Box Plot of Runtime")
plt.ylabel("Runtime (Minutes)")

plt.show()
```
