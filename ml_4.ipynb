```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop_duplicates()
df
```

```python
df.columns.tolist()
```

```python
plt.subplot(1, 2, 1)
if "comic_name" in df.columns:
    df["comic_name"].value_counts().head(10).plot(kind="bar", color="crimson")
    plt.title("Top 10 Marvel Comics by Issue Count")
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Number of Issues")
```

```python
plt.subplot(1, 2, 2)

if "Price" in df.columns:
    
    df["Price_Cleaned"] = pd.to_numeric(
        df["Price"].astype(str).str.replace("$", "", regex=False),
        errors="coerce",
    )

    
    sns.histplot(
        df[df["Price_Cleaned"] > 0]["Price_Cleaned"],
        bins=20,
        kde=True,
        color="gold",
    )
    plt.title("Distribution of Comic Book Prices")
    plt.xlabel("Price ($)")
```
