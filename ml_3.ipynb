```python
import pandas as pd 
import numpy as np
```

```python
ds = pd.read_csv('tested.csv')
ds.head()
```

```python
ds.isnull().count
```

```python
ds_drop = ds.dropna()
ds_drop
```

```python
ds.shape
```

```python
ds_drop.shape
```

```python
ds_fill = ds.copy()

ds_fill["Age"] = ds_fill["Age"].fillna(ds_fill["Age"].mean())
ds_fill["Embarked"] = ds_fill["Embarked"].fillna(ds_fill["Embarked"].mode()[0])

print("Shape after fillna():", ds_fill.shape)
```

```python
ds_fill["Pclass"] = ds_fill["Pclass"].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

ds_fill["Pclass"] = le.fit_transform(ds_fill["Pclass"])
```

```python
ds_encoded = pd.get_dummies(
    ds_fill,
    columns=["Sex", "Embarked"],
    drop_first=True
)
```

```python
print(ds_encoded.select_dtypes(include="object").columns)
```

```python
print(ds_encoded.dtypes)
```
